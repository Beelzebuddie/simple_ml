```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```python
#Data Preparation
##Data separation as X and Y

y = df['logS']
y
```

```python
x = df.drop('logS', axis=1) #drop logS and axis=1 means column
x
```

```python
## Data Splitting into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
# Model Building

## Linear Regression
```

```python
### training the model
from sklearn.linear_model import LinearRegression

lr = LinearRegression() 
lr.fit(x_train, y_train) #training the model
```

```python
### Applying model to make predictions
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
y_lr_train_pred #represents 80% predicted values
```

```python
y_lr_test_pred #represents 20% predicted values
```

```python
## Evaluate model performance

from sklearn.metrics import mean_absolute_error, r2_score

# These 2 blocks are for the training set
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred) 

# These 2 blocks are for the test set
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
# Results
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
# rename the columns
lr_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
lr_results
```

```python
## Random Forest
### training the model
from sklearn.ensemble import RandomForestRegressor # if the Y variable is countable we use RandomForestRegressor. If it is categorical we use RandomForestClassifier

rf = RandomForestRegressor(max_depth=2, random_state=100) # max_depth is the depth of the tree
rf.fit(x_train, y_train) #training the model            
```

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

# These 2 blocks are for the training set
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred) 

# These 2 blocks are for the test set
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
# rename the columns
rf_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
rf_results
```

```python
## Model comparison
# concatenate the results and reset the index
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```

```python
# visualize the results
import matplotlib.pyplot as plt 
import numpy as np # for trend line

plt.figure(figsize=(5, 5)) # set the figure size
plt.scatter(x=y_train, y=y_lr_train_pred, c="#7CAE00", alpha=0.3) # plot the trend line

z = np.polyfit(y_train, y_lr_train_pred, 1) # find the trend line
p = np.poly1d(z) # define the trend line

plt.plot(y_train, p(y_train), '#F8766D') # plot the trend line
plt.ylabel('Predicted LogS') #  label the y axis
plt.xlabel('Experimental LogS') # label the x axis
```
